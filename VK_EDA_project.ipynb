**Virat Kohli** EDA analysis:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/ViratKohli.csv")
df.info()
```

```python
country_counts = df['Against'].value_counts()
country_counts
```

```python
plt.figure(figsize=(12,6))
l = list(country_counts)
country_counts.plot(kind='bar',xlabel='Against',ylabel='No.of Centuries')
for i,v in enumerate(l):
  plt.text(i,v,f'{v}',ha='center',va='bottom')
plt.title('Virat centuries against each team')









```

```python
plt.figure(figsize=(12,6))
Venue_records = df['Venue'].value_counts()
Venue_records[Venue_records>2].plot(kind='bar',color='green',xlabel='Venue',ylabel='No.of centuries')
for i,v in enumerate(list(Venue_records[Venue_records>2])):
  plt.text(i,v,f'{v}',ha='center',va='bottom')
plt.title('Virat kohli outstanding performanes in particular venue')


```

```python
plt.figure(figsize=(14,6))
plt.subplot(1,3,1)
plt.title('Home/Away')
ha=df['H/A'].value_counts()
plt.bar(ha.index,ha.values,color='orange',width=0.4)
plt.xlabel('Home/Away')
plt.ylabel('No.of centuries')
for i,v in enumerate(ha.values):
  plt.text(i,v,f'{v}',ha='center',va='bottom')

plt.subplot(1,3,2)
plt.title('frequency of Man of the Match')
m=df['Man of the Match'].value_counts()
m.plot(kind='bar',xlabel='man of the match')
plt.ylabel('No.of times',labelpad=1)
plt.xticks(rotation=0)
for i,v in enumerate(m.values):
  plt.text(i,v,f'{v}',ha='center',va='bottom')

plt.subplot(1,3,3)
plt.title('As a Captain & Player')
c=df['Captain'].value_counts()
c.plot(kind='bar',xlabel='As a Captain',ylabel='No.of centuries',color='green')
plt.xticks(rotation=0)
for i,v in enumerate(c.values):
  plt.text(i,v,f'{v}',ha='center',va='bottom')




```

```python
plt.figure(figsize= (12,6))
plt.title('Overview Of Result')
plt.subplot(1,2,1)
res=df['Result'].value_counts()
plt.bar(res.index,res.values,width=0.4,color=['green','red','gray','magenta','lightgreen','darkgray'])
plt.xlabel('Result')
plt.ylabel('No.of matches')
for i,v in enumerate(res.values):
  plt.text(i,v,f'{v}',ha='center',va='bottom')




```

```python
from ctypes import alignment
import plotly.express as px
s=px.pie(res,names=res.index,values=res.values,hole=0.5,color_discrete_sequence=['green','red','gray','magenta','lightgreen','darkgray'],title='Win percentage when Virat scores a century')
s.update_traces(pull=[0.1,0,0,0,0,0],rotation=140)
s.update_layout(title_x=0.5)
```

```python
b=df['Batting Order'].value_counts()
b.plot(kind='bar',xlabel='When played at number',ylabel='No.of centuries',color='magenta')
plt.xticks(rotation=0)
for i,v in enumerate(b.values):
  plt.text(i,v,f'{v}',ha='center',va='bottom')
```
